## 2. x轴和y轴名称

xlabel() ylabel()

```python
# 创建x为-10到10的整数
x = np.arange(-10,10)
# y是x的平方
y = x ** 2
# 设置标题
plt.title('y=x^2 x取值范围为:[-10,10)')
# 设置x轴名称
plt.xlabel("x 轴")
# 设置y轴名称
plt.ylabel("y 轴")

plt.plot(x, y)
```

#### 对于上面的图,如果觉得字体偏小或线条太细,可以设置标签文字大小和线条粗细

fontsize参数: 设置文字大小

linewidth参数: 设置线条

```python
# 创建x为-10到10的整数
x = np.arange(-50,51)
# y是x的平方
y = x ** 2
# 设置标题
plt.title('y=x^2 x取值范围为:[-10,10)',fontsize=20)
# 设置x轴名称 ,值不能写"12px"
plt.xlabel("x 轴",fontsize=11)

# 设置y轴名称
plt.ylabel("y 轴")
plt.plot(x, y,linewidth=5)
plt.plot(x, y,linewidth=1)
plt.plot(x, y,linewidth=3)
x_titcks=[-50,-40,-30,-20,-10,0,10,20,30,40,50]
x_lables=['%s度'%i for i in x_titcks]
plt.xticks(x_titcks,x_lables)

```

```python
# 设置线条宽度
plt.plot(x, y,linewidth=1)
plt.plot(x, y,linewidth=3)
```

#### 运行上面代码,除了线条变化外,大家有没有发现图像有什么变化?

```python
# 我们可以在一张图片中绘制多个线条
# 创建x为-10到10的整数
x = np.arange(-10,10)
# y1是x的平方
y1 = x ** 2
# y2等于x
y2 = x
#y3等于x的立方
y3=x ** 3
# 设置标题
plt.title('y=x^2 x取值范围为:[-10,10)',fontsize=16)
# 设置x轴名称
plt.xlabel("x 轴",fontsize=12)
# 设置y轴名称
plt.ylabel("y 轴")
# 绘图线条1
plt.plot(x, y1)
# 绘图线条2
plt.plot(x, y2)
#绘图线条3
plt.plot(x,y3)
```

### 3. 设置x轴和y轴的刻度

`matplotlib.pyplot.xticks(ticks=None, labels=None, **kwargs)`

`matplotlib.pyplot.yticks(ticks=None, labels=None, **kwargs)`

- ticks: 此参数是xtick位置的列表。和一个可选参数。如果将一个空列表作为参数传递，则它将删除所有xticks
- label: 此参数包含放置在给定刻度线位置的标签。它是一个可选参数。，
- **kwargs:此参数是文本属性，用于控制标签的外观
 - rotation:旋转角度 如:rotation=45
 - color:颜色  如:color="red"

> <font color="red">xticks到底有什么用，其实就是想把坐标轴变成自己想要的样子</font>

```python
# 日期
times = np.arange(1990,2020).astype(np.str_)
# 销量
sales = np.random.randint(500,2000,size=len(times))
# 绘图
plt.xticks(range(0,len(times),2))
plt.plot(times,sales)
```

```python
# 每个时间点的销量绘图,如果x轴的数据是字符串,会将数据全部展示
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/9','2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']
# 随机出销量
sales =np.random.randint(500,2000,size=len(times))
# 绘制图形
plt.plot(times,sales)
```

```python
# 以上图形x轴存在的问题?

# 如果想只显示部分时间,或者按照某个规则展示,如何处理,这个时候就用到xticks
plt.xticks(range(0,len(times),3))#跳过了3个时间
plt.plot(times,sales)
```

```python
# 时间虽然少了,可不可以让横坐标中的字体倾斜,就可以放下了.rotation:旋转角度 如:rotation=45就是旋转45°
plt.xticks(range(0,len(times),2),rotation=45)#跳过了3个时间
plt.plot(times,sales)
```

```python
# 时间虽然少了,可不可以让横坐标中的字体倾斜,就可以放下了.
# xticks 第一个参数中元素的值,代表原始数据的索引
plt.xticks(range(0,len(times),2),labels=[1,2,3,4,5,6,7,8], rotation=45, color='red')
plt.plot(times,sales)
```

```python
# 时间虽然少了,可不可以让横坐标中的字体倾斜,就可以放下了.
# xticks 第一个参数中元素的值,代表原始数据的索引
#由于labels标签要和跳过的时间要匹配
plt.xticks(range(0,len(times),2),labels=['日期:%s'%i for i in times], rotation=45, color='green')
plt.plot(times,sales)
```

### 5.图例 legend()

图例是集中于地图一角或一侧的地图上各种符号和颜色所代表内容与指标的说明，有助于更好的认识地图。

<img src="images/image-20220316222353788.png" style="width:400px"/>


```python
# 每个时间点的销量绘图
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/9','2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']
# 随机出收入
income =np.random.randint(500,2000,size=len(times))
# 支出
expenses =np.random.randint(300,1500,size=len(times))
# 绘制图形
plt.xticks(range(1,len(times),2),rotation=45,color='green')
# 注意,在使用图例前为每个图形设置label参数
plt.plot(times,income,label="收入")
plt.plot(times,expenses,label="支出")
# 默认会使用每个图形的label值作为图例中的说明
plt.legend()
```

### 显示每条数据的值 x,y值的位置

`plt.text(x,y,  string,  fontsize=15,  verticalalignment="top",  horizontalalignment="right") `


- x,y:表示坐标值上的值
- string:表示说明文字
- fontsize:表示字体大小
- verticalalignment：(va)垂直对齐方式 ，参数：[ ‘center’ | ‘top’ | ‘bottom’ | ‘baseline’ ]
- horizontalalignment：(ha)水平对齐方式 ，参数：[ ‘center’ | ‘right’ | ‘left’ ]




```python
# 每个时间点的销量绘图
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/9',
'2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']
print(times)
# 随机出收入
income =np.random.randint(500,2000,size=len(times))
# 支出
expenses =np.random.randint(300,1500,size=len(times))

# 绘制图形
plt.xticks(range(1,len(times),2),rotation=45)
# 注意,在使用图例前为每个图形设置label参数
plt.plot(times,income,label="收入")
plt.plot(times,expenses,label="支出")
# 默认会使用每个图形的label值作为图例中的说明
plt.legend(loc='upper center')
for x,y in zip(times,income):
    plt.text(x,y,y)

#for a,b in zip(times,expenses):
#    plt.text(a,b,b)
```
